```python
# Graph
# Adjacency List for a unweighted graph
class Edge:
  def __init__(self, source, destination):
    self.source = source
    self.destination = destination
    self.next = None

import numpy as np
def adjacency_list():
  adj_list = [None] * 5

  # For A
  adj_list[0] = Edge("A", "C")
  adj_list[0].next = Edge("A", "E")

  # For B
  adj_list[1] = Edge("B", "A")
  adj_list[1].next = Edge("B", "C")
  adj_list[1].next.next = Edge("B", "E")

  # For C index 2 will be empty in this case None

  # For D
  adj_list[3] = Edge("D", "B")
  adj_list[3].next = Edge("D", "C")
  adj_list[3].next.next = Edge("D", "E")
  adj_list[3].next.next.next = Edge("D", "D")

  # For E
  adj_list[4] = Edge("E", "B")
  adj_list[4].next = Edge("E", "C")
  adj_list[4].next.next = Edge("E", "A")

  return adj_list

graph_list = adjacency_list()
vertex = ['A', 'B', 'C', 'D', 'E']

for i in range(len(graph_list)):
  print(f"Printing for Node {vertex[i]}: ", end="")
  temp = graph_list[i]
  while temp != None:
    if temp.next:
        print(f"{temp.destination}", end=" --> ")
    else:
        print(f"{temp.destination}")
    temp = temp.next
  print()
```
