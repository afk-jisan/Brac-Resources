```python
def find_path(root, key):
    if root is None:
        return None  # Return None if no path is found

    if root.elem == key:
        return [root.elem]  # Return the path as a list containing the current node

    # Recursively search in the left subtree if the key is smaller
    if key < root.elem:
        left_path = find_path(root.left, key)
        if left_path is not None:  # Check if a valid path was found in the left subtree
            return [root.elem] + left_path

    # Recursively search in the right subtree if the key is larger
    if key > root.elem:
        right_path = find_path(root.right, key)
        if right_path is not None:  # Check if a valid path was found in the right subtree
            return [root.elem] + right_path

    return None  # Return None if no path is found


# DRIVER CODE
root = BTNode(30)
root.left = BTNode(10)
root.right = BTNode(40)
root.left.left = BTNode(3)
root.left.right = BTNode(15)
root.right.left = BTNode(35)
root.right.right = BTNode(55)

# Test cases
result1 = find_path(root, 15)
print(result1 if result1 is not None else "No Path Found")  # This should print [30, 10, 15]

result2 = find_path(root, 50)
print(result2 if result2 is not None else "No Path Found")  # This should print "No Path Found"
```

```python
def inorder_predecessor(root, x):
    """
    Finds the inorder predecessor of a given node `x` in a Binary Search Tree.
    :param root: Root node of the BST.
    :param x: Node whose inorder predecessor is to be found.
    :return: Inorder predecessor node or None if no predecessor exists.
    """
    pred = None
    temp = root

    while temp:
        if x.elem <= temp.elem:
            # Move to the left subtree
            temp = temp.left
        else:
            # Update predecessor and move to the right subtree
            pred = temp
            temp = temp.right

    return pred



#DRIVER CODE
root = BTNode(20)
n1 = BTNode(8)
n2 = BTNode(22)
n3 = BTNode(4)
n4 = BTNode(12)
n5 = BTNode(10)
n6 = BTNode(14)

root.left = n1
root.right = n2

n1.left = n3
n1.right = n4

n4.left = n5
n4.right = n6

print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root) #Given Tree Inorder Traversal:  4 8 10 12 14 20 22
print()

x = root
print(f'Inorder predecessor of node {x.elem}: {inorder_predecessor(root, x).elem}') #Inorder predecessor of node 20: 14
x = root.left.right.right
print(f'Inorder predecessor of node {x.elem}: {inorder_predecessor(root, x).elem}') #Inorder predecessor of node 14: 12
x = root.left.right.left
print(f'Inorder predecessor of node {x.elem}: {inorder_predecessor(root, x).elem}') #Inorder predecessor of node 10: 8

#Leaf node is the right child? --> immediate parent
x = root.left.right
print(f'Inorder predecessor of node {x.elem}: {inorder_predecessor(root, x).elem}') #Inorder predecessor of node 10: 8

x = root.right
print(f'Inorder predecessor of node {x.elem}: {inorder_predecessor(root, x).elem}') #Inorder predecessor of node 20: 14
```
