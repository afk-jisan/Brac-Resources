**Assignment Part 1: Doubly Linked List**

For Assignment Part 1, you should write your full code in the following cells along with your driver codes on your own.

```python
#Assignment Part 1

class Patient:
  #write a constructor
  def __init__(self, id, name, age, bloodgroup, next, prev):
    self.id = id
    self.name = name
    self.age = age
    self.bloodgroup = bloodgroup
    self.next = next
    self.prev = prev
```

```python
class WRM:

    def __init__(self):
        #Creating the dummy head
        self.dh = Patient(None,None,None,None,None,None)
        self.dh.next = self.dh
        self.dh.prev = self.dh


    def registerPatient(self,id, name, age, bloodgroup):
        new = Patient(id, name, age, bloodgroup, self.dh, self.dh.prev)
        self.dh.prev.next = new
        self.dh.prev = new



    def servePatient(self):
        if self.dh.next != self.dh:
            n1 = self.dh.next
            n2 = n1.next
            print(f"{n1.name} got served")
            self.dh.next = n2
            n2.prev = self.dh
        else:
            print("Found no patient to serve")


    def showAllPatient(self):
        if self.dh.next != self.dh:
            temp = self.dh.next
            while temp != self.dh:
                print(f"{temp.id} - {temp.name}")
                temp = temp.next
        else:
            print(f"No patient found")



    def canDoctorGoHome(self):
        if self.dh.next == self.dh:
            return True
        else:
            return False


    def cancelAll(self):
        self.dh.next = self.dh
        self.dh.prev = self.dh



    def ReverseTheLine(self):
        temp = self.dh.next
        firstNode = self.dh
        while temp != self.dh:
            newhead = temp.next

            temp.next = firstNode
            firstNode.prev = temp
            self.dh.next = temp
            temp.prev = self.dh

            temp = newhead
            firstNode = self.dh.next
```

```python
#Write a Tester Code in this cell
print("**Welcome to Waiting Room Management System**")

linkedList = WRM()
while True:
    print("___________________________________________________________________________________________________________________________________________ ")
    print(' 1. Add Patient\n 2. Serve Patient\n 3. Show All patients\n 4. Can Doctor go Home?\n 5. Cancel all Appointment\n 6. ReverseTheLine\n 7. Exit')
    userInput = int(input("Enter your choosen option no : "))

    if userInput == 1:
        id = int(input('Enter id: '))
        name = input('Enter name: ')
        age = int(input('Enter age: '))
        bloodgroup = input('Enter blood group: ')
        linkedList.registerPatient(id, name, age, bloodgroup)
        print("Success registering a new patient")

    elif userInput == 2:
        linkedList.servePatient()
    elif userInput == 3:
        linkedList.showAllPatient()
    elif userInput == 4:
        if linkedList.canDoctorGoHome():
            print("Yes")
        else:
            print("No")


    elif userInput == 5:
        if linkedList.canDoctorGoHome():
            print("No Appointment to cancel")
        else:
            print("Successfully Cancelled all Appointment")
            linkedList.cancelAll()

    elif userInput == 6:
        if linkedList.canDoctorGoHome():
            print("No Line to reverse")
        else:
            linkedList.ReverseTheLine()
            print("Successfully Reversed The Line")


    elif userInput == 7:
        break

```

**Assignment Part 2: Stack**

Linked List based Stack is implemented in the following cell.

```python
class Node:
  def __init__(self,elem=None,next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    e = self.__top
    self.__top = self.__top.next
    return e.elem

  def peek(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
```

Task 1: Diamond Count

```python
def diamond_count(stack,string):
    count = 0
    for i in string:
        if i == '<':
            stack.push(i)

        elif stack.peek() == '<' and i == '>':
            stack.pop()
            count += 1
    return count





print('Test 01')
stack = Stack()
string = '<..><.<..>> '
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')


print('Test 02')
stack = Stack()
string = '<<<..<......<<<<....>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 1
unittest.output_test(returned_value, 1)
print('-----------------------------------------')


print('Test 03')
stack = Stack()
string = '>>><...<<..>>...>...>>>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')
```

Task 2: Tower of Blocks

```python
def remove_block(stack, n):
    count = 1
    temp = Stack()
    while stack.peek() != None:
        if count == n:
            stack.pop()
        else:
            item = stack.pop()
            temp.push(item)
        count += 1

    while temp.peek() != None:
        st.push(temp.pop())




print('Test 01')
st = Stack()
st.push(4)
st.push(19)
st.push(23)
st.push(17)
st.push(5)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,2)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()

print('Test 02')
st = Stack()
st.push(73)
st.push(85)
st.push(15)
st.push(41)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,3)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()
```

Task 3: Stack Reverse

```python
def conditional_reverse(stack):
    temp = Stack()
    while stack.peek() != None:
        item = stack.pop()

        if item == stack.peek():
            temp.push(item)
            stack.pop()
        else:
            temp.push(item)
    return temp



print('Test 01')
st=Stack()
st.push(10)
st.push(10)
st.push(20)
st.push(20)
st.push(30)
st.push(10)
st.push(50)
print('Stack:')
print_stack(st)
print('------')
reversed_stack=conditional_reverse(st)
print('Conditional Reversed Stack:')
print_stack(reversed_stack) # This stack contains 50, 10, 30, 20, 10 in this order whereas top element should be 10
print('------')
```
