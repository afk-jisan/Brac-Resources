```python
#Task - 2
import numpy as np

class MaxHeap:
    def __init__(self, capacity):
        self.__heap = np.zeros(capacity, dtype=int)
        self.__size = 0

    def insert(self, item):
        if self.__size == len(self.__heap)-1:
            print("No space left in Heap")
        else:
            self.__size += 1
            self.__heap[self.__size] = item
            self.__swim()


    def __swim(self):
        child = self.__size
        parent = child // 2

        while parent > 0 and self.__heap[child] >= self.__heap[parent]:
            temp = self.__heap[parent]
            self.__heap[parent] = self.__heap[child]
            self.__heap[child] = temp

            child = parent
            parent = child // 2


    def extractMax(self):
        if self.__size == 0:
            return "No More item in heap"

        deletedVal = self.__heap[1]
        self.__heap[1] = self.__heap[self.__size]
        self.__heap[self.__size] = deletedVal
        self.__size -= 1
        self.__sink()
        return deletedVal

    def __sink(self):
        parent = 1
        left = parent * 2
        right = (parent * 2) + 1
        while left <= self.__size and right <= self.__size:
            if self.__heap[parent] <= self.__heap[left] or self.__heap[parent] <= self.__heap[right]:
                if self.__heap[left] > self.__heap[right]:
                    temp = self.__heap[parent]
                    self.__heap[parent] = self.__heap[left]
                    self.__heap[left] = temp

                    parent = left
                else:
                    temp = self.__heap[parent]
                    self.__heap[parent] = self.__heap[right]
                    self.__heap[right] = temp

                    parent = right

            elif self.__heap[parent] < self.__heap[left]:
                temp = self.__heap[parent]
                self.__heap[parent] = self.__heap[left]
                self.__heap[left] = temp

                parent = left
            elif self.__heap[parent] < self.__heap[right]:
                temp = self.__heap[parent]
                self.__heap[parent] = self.__heap[right]
                self.__heap[right] = temp

                parent = right
            else:
                break

            left = parent * 2
            right = (parent * 2) + 1

        if left <= self.__size and self.__heap[parent] < self.__heap[left]:
            temp = self.__heap[parent]
            self.__heap[parent] = self.__heap[left]
            self.__heap[left] = temp



    def sort(self):
        copy_heap = np.array(self.__heap)
        copy_size = self.__size
        sorted_arr = np.zeros(copy_size, dtype=int)

        for i in range(self.__size) :
            sorted_arr[i] = self.extractMax()

        self.__heap = copy_heap
        self.__size = copy_size
        return sorted_arr



    def show_array(self):
        print(f"Max Heap [size: {self.__size}]: {self.__heap[1:self.__size+1]}")
        #print(f"Max Heap [size: {self.__size}]: {self.__heap}")



val = np.array([50, 30, 40, 10, 20, 15])
max_heap = MaxHeap(len(val))
print("Max Heap Insertion :")
for i in range(1, len(val)):
  max_heap.insert(val[i])
  max_heap.show_array()


print()
print("Max Heap Deletion :")
while True:
  if max_heap.extractMax() != "No More item in heap":
    max_heap.show_array()
  else:
    break


print()
print("Inserting values into Max Heap again to demonstrate sort method :")
for i in range(1, len(val)):
  max_heap.insert(val[i])
  max_heap.show_array()
print()
print(f"Sorted array: {max_heap.sort()}")
```

```python
#Task - 3
import numpy as np

def machines(tasks, m):
    min_heap_loads = MinHeap(m+1)

    for i in range(m):               #for i in range(1,m+1):
        min_heap_loads.insert(0)

    for i in range(len(tasks)):
        extracted = min_heap_loads.extractMin()
        total = tasks[i] + extracted
        min_heap_loads.insert(total)
    return min_heap_loads.returnHeap()


tasks = np.array([2, 4, 7, 1, 6])
print(machines(tasks, m = 4))
```
